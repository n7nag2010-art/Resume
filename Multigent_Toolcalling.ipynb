```python
def add(a: float, b: float) -> str:
    return str(a + b)

def subtract(a: float, b: float) -> str:
    return str(a - b)

def multiply(a: float, b: float) -> str:
    return str(a * b)

def divide(a: float, b: float) -> str:
    if b == 0:
        return "Error: Division by zero"
    return str(a / b)
```

```python
from langchain_core.tools import Tool

math_tools = [
    Tool(
        name="add",
        func=add,
        description="Add two numbers. Input: a, b"
    ),
    Tool(
        name="subtract",
        func=subtract,
        description="Subtract b from a. Input: a, b"
    ),
    Tool(
        name="multiply",
        func=multiply,
        description="Multiply two numbers. Input: a, b"
    ),
    Tool(
        name="divide",
        func=divide,
        description="Divide a by b. Input: a, b"
    ),
]
```

```python
from langchain_groq import ChatGroq

from dotenv import load_dotenv
import os


load_dotenv()  # reads .env file

groq_api_key = os.getenv("GROQ_API_KEY")

print(groq_api_key)

llm = ChatGroq(
    api_key=groq_api_key,
    model_name="openai/gpt-oss-20b",
    temperature=0
)

llm_with_tools = llm.bind_tools(math_tools)
```

```python
messages = [
    {"role": "user", "content": "What is 25 * 4 +10?"}
]

messages = [
    {"role": "user", "content": "What is 25muliply with 25 and add 10 ?"}
]

response = llm_with_tools.invoke(messages)
print(response.content)
```
